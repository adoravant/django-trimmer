```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(open("index.html"))
            
```

```python
links = (lin for lin in soup.find_all("a"))
while True:
    link = next(links)
    try:
        print(link, link["href"]")
        tell = input("what, ")
        if tell == "x":
            link.extract()
        elif tell == "move":
            continue
        elif tell == "break":
            break
    except:
        pass
```
